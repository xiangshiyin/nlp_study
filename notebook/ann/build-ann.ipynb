## Part 4: Evaluate the model

#### evaluate the ANN (k-fold cross validation)

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense

def build_classifier():
    classifier = Sequential() ## define layers step-by-step later
    classifier.add(layer=Dense(
        units=6,
        kernel_initializer='uniform',
        activation='relu',
        input_dim=11
    )) ## the 1st hidden layer

    classifier.add(layer=Dense(
        units=6,
        kernel_initializer='uniform',
        activation='relu'
    )) ## the 2nd hidden layer

    classifier.add(layer=Dense(
        units=1,
        kernel_initializer='uniform',
        activation='sigmoid'
    )) ## the output layer

    classifier.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return classifier

classifier = KerasClassifier(
    build_fn=build_classifier,
    batch_size=10,
    epochs=100
)
accuracies = cross_val_score(
    estimator=classifier,
    X=X_train,
    y=y_train,
    cv=10,
    n_jobs=-1
)
```

#### Dropout regularization to reduce overfitting if needed

```python
from keras.layers import Dropout
```

```python
def build_classifier():
    classifier = Sequential() ## define layers step-by-step later
    classifier.add(layer=Dense(
        units=6,
        kernel_initializer='uniform',
        activation='relu',
        input_dim=11
    )) ## the 1st hidden layer
    classifier.add(Dropout(rate=0.1))

    classifier.add(layer=Dense(
        units=6,
        kernel_initializer='uniform',
        activation='relu'
    )) ## the 2nd hidden layer
    classifier.add(Dropout(rate=0.1))

    classifier.add(layer=Dense(
        units=1,
        kernel_initializer='uniform',
        activation='sigmoid'
    )) ## the output layer

    classifier.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return classifier

classifier = KerasClassifier(
    build_fn=build_classifier,
    batch_size=10,
    epochs=100
)
accuracies = cross_val_score(
    estimator=classifier,
    X=X_train,
    y=y_train,
    cv=10,
    n_jobs=-1
)
mean = accuracies.mean()
variance = accuracies.std()
mean, variance
```

#### Tune the ANN (grid search)

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense

def build_classifier(optimizer):
    classifier = Sequential() ## define layers step-by-step later
    classifier.add(layer=Dense(
        units=6,
        kernel_initializer='uniform',
        activation='relu',
        input_dim=11
    )) ## the 1st hidden layer

    classifier.add(layer=Dense(
        units=6,
        kernel_initializer='uniform',
        activation='relu'
    )) ## the 2nd hidden layer

    classifier.add(layer=Dense(
        units=1,
        kernel_initializer='uniform',
        activation='sigmoid'
    )) ## the output layer

    classifier.compile(
        optimizer=optimizer,
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return classifier

classifier = KerasClassifier(
    build_fn=build_classifier
#     ,
#     batch_size=10,
#     epochs=100 ##(will be defined in grid search)
)

parameters = {
    'batch_size':[25,32],
    'epochs':[100,500],
    'optimizer':['adam','rmsprop']
}

grid_search = GridSearchCV(
    estimator=classifier,
    param_grid=parameters,
    scoring='accuracy',
    cv=10
)

grid_search = grid_search.fit(X_train, y_train)
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
```
